## Problem 1
A system of equation $Ax = b$ is given as
$$\begin{bmatrix}
2.01 & 1.99\\
1.99 & 2.01
\end{bmatrix}
x = 
\begin{bmatrix}
4\\
4
\end{bmatrix}.$$
The Solution to this system of equations is 
$$x = 
\begin{bmatrix}
1\\
1
\end{bmatrix},$$
and one approximate solution is 
$$\tilde{x} = 
\begin{bmatrix}
2\\
0
\end{bmatrix}.$$
(i) Compute the error $e = \tilde{x} - x$.

(ii) Compute the residual $r = A\tilde{x} - b$.

(iii) Using the $l_{\infty}$ norm, compute the relative error $\tfrac{\|e\|_{\infty}}{\|x\|_{\infty}}$.

(iv) Using the $l_{\infty}$ norm, compute the condition number $\kappa_{\infty}$. 

(v) Using the $l_{\infty}$ norm, compute the relative residual $\tfrac{\|r\|_{\infty}}{\|b\|_{\infty}}$.

(vi) Compute the product of the condition number and the relative residual.

(vii) Compare the relative error to the product of the condition number and the relative residual.

**Solution: ** 

(i) The error is 
$$e = \tilde{x} - x = \begin{bmatrix}
1\\
-1
\end{bmatrix}$$.

(ii) The residual is:
$$r = A\tilde{x} - b = \begin{bmatrix}
2.01 & 1.99\\
1.99 & 2.01
\end{bmatrix}\begin{bmatrix}
2\\
0
\end{bmatrix} - \begin{bmatrix}
4\\
4
\end{bmatrix} = 
\begin{bmatrix}
0.02\\
-0.02
\end{bmatrix}$$

(iii) The relative error is:
$$\frac{\|e\|_{\infty}}{\|x\|_{\infty}} = \frac{\max_{i} |e_i|}{\max_{i} |x_i|} = \frac{1}{1} = 1$$

(iv) The coefficient matrix A and inverse of matrix A are given below:
$$A = \begin{bmatrix}
2.01 & 1.99\\
1.99 & 2.01
\end{bmatrix},\quad
A^{-1} = \begin{bmatrix}
25.1250 & -24.8750\\
-24.8750 & 25.1250
\end{bmatrix}$$
Therefore, the condition number is:
$$\kappa_{\infty} = \|A\|_{\infty}\cdot\|A^{-1}\|_{\infty} = 4\times 50 = 200$$

(v) The relative residual is:
$$\frac{\|r\|_{\infty}}{\|b\|_{\infty}} = \frac{0.02}{4} = 0.005$$

(vi) The product of the condition number and the relative residual:
$$\kappa_{\infty}\times \frac{\|r\|_{\infty}}{\|b\|_{\infty}} = 200 \times 0.005 = 1$$

(vii) The relative error are equal to the product of the condition number and the relative residual.

(v) Theorem: Assume $Ax = b$ and $\hat{A}\hat{x} = \hat{b}$. If $A$ is nonsingular and $\|\hat{A} - A\| < \frac{1}{\|A^{-1}\|}$, there holds
$$\frac{\|\hat{x} - x\|}{\|x\|} \leq \frac{cond(A)}{1-cond(A)\frac{\|\hat{A} - A\|}{\|A\|}}(\frac{\|\hat{b}-b\|}{\|b\|} + \frac{\|\hat{A} - A\|}{\|A\|})$$

$$A(\hat{x} - x) = \hat{b} - b - E\hat{x}$$
$$\|\hat{x} - x\| \leq \|A^{-1}\|(\|\hat{b} - b\| + \|E\|\cdot\|\hat{x}\|)

Therefore,
$$(1 - \|A^{-1}\|\cdot\|E\|)\|\hat{x} - x\| \leq \|A^{-1}\|(\|\hat{b}-b\| + \|E\|\|x\|)$$
Since $1 \leq \tfrac{\|A\|\|x\|}{\|b\|}$ and $\|\hat{A} - A\| < \frac{1}{\|A^{-1}\|}$,
$$(1 - \|A^{-1}\|\cdot\|E\|)\|\hat{x} - x\| \leq \|A^{-1}\|(\frac{\|\hat{b}-b\|}{\|b\|}\cdot \|A\|\|x\| + \|E\|\|x\|)$$
Hence, the theorem is proved:
$$\frac{\|\hat{x} - x\|}{\|x\|} \leq \frac{cond(A)}{1-cond(A)\frac{\|\hat{A} - A\|}{\|A\|}}(\frac{\|\hat{b}-b\|}{\|b\|} + \frac{\|\hat{A} - A\|}{\|A\|})$$

(i) The codes for solving this system of linear equations using Gaussian Elimination with Partial Pivoting is shown below:

```python
import numpy as np
from matplotlib import pyplot as plt
# Implementation of Gaussian Elimination with Partial Pivoting
# This algorithm computes a unique solution of the system, or 
# indicate that system has no unique solution
# Partial Pivoting when Akk is less than 0.1% of maximum in cl
def gaussElim(A, b):
    n = len(b)
    for k in range(0,n-1):
        '''
        maximum = abs(A[k,k])
        r = k
        for i in range(k+1,n):
            if abs(A[i,k]) > maximum:
                maximum = abs(A[i,k])
                r = i
        '''
        r = abs(A[k:,k]).argmax() + k
        if A[r,k] == 0.0:
            print("The system does not have unique solution.")
            break
        else:
            '''
            tmp = A[r,:].copy()
            tmpb = b[r].copy()
            A[r,:] = A[k,:]
            b[r] = b[k]
            A[k,:] = tmp
            b[k] = tmpb
            '''
            A[[k,r]] = A[[r,k]]
            b[[k,r]] = b[[r,k]]
        for i in range(k+1,n):
            lam = A[i,k]/A[k,k]
            A[i,k+1:n] = A[i,k+1:n] - lam*A[k,k+1:n]
            b[i] = b[i] - lam*b[k]
    b[n-1] = b[n-1]/A[n-1,n-1]
    for k in range(n-2,-1,-1):
        b[k] = (b[k] - A[k,k+1:n].dot(b[k+1:n]))/A[k,k]
    return b

A = np.array([[-9.0, 11.0, -21.0, 63.0, -252.0],\
              [70.0, -69.0, 141.0, -421.0, 1684.0], \
              [-575.0, 575.0, -1149.0, 3451.0, -13801.0],\
              [3891.0, -3891.0, 7782.0, -23345.0, 93365.0],\
              [1024.0, -1024.0, 2048.0, -6144.0, 24572.0]])
b = np.array([-356.0, 2385.0, -19551.0, 132274.0, 34812.0])
# !!note function paremeters are passed by reference here!!
x = gaussElim(A.copy(),b.copy())
print("***ME 581 Homework 2 Print Out***\n")
print("The solution from Gaussian Elimination with Partial Pivoting is ", x)
```
